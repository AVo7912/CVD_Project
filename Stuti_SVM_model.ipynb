```python
# heart_cleaned.csv is in the resource folder
data_path = '/content/drive/My Drive/Project 4 Final Bootcamp/Resources/heart_cleaned.csv'

# Load data using pandas
df_heart = pd.read_csv(data_path)
```

```python
# Get the target variable
target = df_heart["Heart_Disease"]
```

```python
# Get the features.
heart_data = df_heart.drop("Heart_Disease", axis=1)
feature_names = heart_data.columns
feature_names
```

```python
# Split data into training and testing
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(heart_data, target, random_state=65)
```

```python
from sklearn.svm import SVC  # Corrected import statement

# Create the SVM model with a rdf kernel
model = SVC(kernel='rbf')

# Fit the model to your training data
model.fit(X_train, y_train)
```

```python
#making predictions with test dataset
predictions = model.predict(X_test)
predictions
```

```python
# Model Accuracy
print('Test Acc: %.3f' % model.score(X_test, y_test))
```

```python
# Calculate the classification report
from sklearn.metrics import classification_report
predictions = model.predict(X_test)
print(classification_report(y_test, predictions))
```
